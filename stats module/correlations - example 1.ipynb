# CORRELATING POWERLIFTING RECORDS

```python
# import packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# import dataset

```python
lifting=pd.read_csv('/Users/hlind/Downloads/power_lifting.csv')
```

```python
lifting.head()
```

# correlating age with best squat

```python
lifting['Age'].corr(lifting['BestSquatKg'])
```

# correlating age with best deadlift

```python
lifting['Age'].corr(lifting['BestDeadliftKg'])
```

# creating matrix

```python
lifts=lifting.drop(['Name', 'Sex', 'Equipment', 'Division', 'Squat4Kg', 'Bench4Kg', 'Deadlift4Kg', 'MeetID'], axis=1)
```

```python
# veryfing work
lifts.head()
```

# correlating matrix

```python
lifts.corr(method='pearson')
```

```python
# adding pizzazz
lifts.corr(method='pearson').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

```python
# best squat, bench, and deadlift all correlate with age
```
