```python
import pandas as pd
```

```python
#import data
eatingHabits=pd.read_csv('/Users/hlind/Downloads/Eating_Habits.csv')
```

```python
eatingHabits.Activity.value_counts()
```

```python
#recoding ativity column
def activity(series):
    if series=="Eating fruit":
        return 0
    if series=="Eating candy, chocolate bars":
        return 1
    if series=="Eating french fries":
        return 0
    if series=="Eating hamburgers, hot dogs or sausages":
        return 0
    if series=="Eating whole wheat or rye bread":
        return 0
    if series=="Drinking soft drinks, cola or other drinks with sugar":
        return 0
    if series=="Drinking coffee":
        return 0
    if series=="Eating raw vegetables":
        return 0
    if series=="Eating potato chips, crisps":
        return 0
    if series=="Drinking low fat milk":
        return 0
    if series=="Drinking whole milk":
        return 0
    if series=="Eating peanuts":
        return 0
```

```python
eatingHabits['ActivityR']=eatingHabits['Activity'].apply(activity)
```

```python
#view data
eatingHabits.head()
```

```python
#recode frequency column from text to nums
eatingHabits.Frequency.value_counts()
```

```python
#creating dictionary
Frequency={"Never":0, "Seldom":1, "At least once a week":2, "Once a day":3, "More than once a day":4}
```

```python
eatingHabits.replace(Frequency, inplace=True)
```

```python
#checking data
eatingHabits.head()
```

```python
#recoding sex column from text to nums
eatingHabits.Sex.value_counts()
```

```python
#creating dictionary
Sex={"Males":0, "Females":1}
```

```python
eatingHabits.replace(Sex, inplace=True)
```

```python
#checking data
eatingHabits.head()
```
